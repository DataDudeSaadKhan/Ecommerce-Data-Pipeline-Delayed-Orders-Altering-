```python
import pandas as pd
from datetime import timedelta
import random

# Number of records
N = 100  

# Generate Orders
orders = pd.DataFrame({
    'order_id': [f'O{i}' for i in range(1, N+1)],
    'customer_id': [f'C{random.randint(1, 5)}' for _ in range(N)],
    'order_date': pd.date_range(start='2024-01-01', periods=N, freq='D')
})
orders.to_csv('orders.csv', index=False)

# Generate Customers
customers = pd.DataFrame({
    'customer_id': [f'C{i}' for i in range(1, 6)],
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'email': [f'user{i}@example.com' for i in range(1, 6)]
})
customers.to_csv('customers.csv', index=False)

# Generate Shipments
statuses = ['shipped', 'delivered']
shipped_dates = pd.date_range(start='2024-01-02', periods=N, freq='D')

# Create variable delivery times
delivered_dates = []
for ship_date in shipped_dates:
    # Random delivery delay between 1 and 7 days
    delay_days = random.randint(1, 7)
    delivered_dates.append(ship_date + timedelta(days=delay_days))

shipments = pd.DataFrame({
    'shipment_id': [f'S{i}' for i in range(1, N+1)],
    'order_id': orders['order_id'],
    'status': [random.choice(statuses) for _ in range(N)],
    'shipped_at': shipped_dates,
    'delivered_at': delivered_dates
})
shipments.to_csv('shipments.csv', index=False)
```
